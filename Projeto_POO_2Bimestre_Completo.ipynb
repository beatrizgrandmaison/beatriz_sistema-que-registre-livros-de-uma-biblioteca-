## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
# Classe Livro com m√©todos adicionais
class Livro:
    # Construtor para inicializar os atributos do livro
    def __init__(self, titulo, autor, ano_publicacao):
        self.titulo = titulo
        self.autor = autor
        self.ano_publicacao = ano_publicacao
        self.emprestado = False  # Atributo para verificar se o livro est√° emprestado

    # M√©todo para exibir informa√ß√µes sobre o livro
    def exibir_dados(self):
        print(f"T√≠tulo: {self.titulo}")
        print(f"Autor: {self.autor}")
        print(f"Ano de Publica√ß√£o: {self.ano_publicacao}")
        print(f"Emprestado: {'Sim' if self.emprestado else 'N√£o'}")
        print("-" * 30)
    
    # M√©todo para marcar o livro como emprestado
    def emprestar(self):
        if not self.emprestado:
            self.emprestado = True
            print(f"O livro '{self.titulo}' foi emprestado.")
        else:
            print(f"O livro '{self.titulo}' j√° est√° emprestado.")

    # M√©todo para devolver o livro
    def devolver(self):
        if self.emprestado:
            self.emprestado = False
            print(f"O livro '{self.titulo}' foi devolvido.")
        else:
            print(f"O livro '{self.titulo}' n√£o estava emprestado.")

# Subclasse LivroDigital, herdando de Livro
class LivroDigital(Livro):
    def __init__(self, titulo, autor, ano_publicacao, formato_arquivo):
        # Usando super() para inicializar a parte da classe pai
        super().__init__(titulo, autor, ano_publicacao)
        self.formato_arquivo = formato_arquivo  # Novo atributo exclusivo para livros digitais

    # M√©todo exclusivo da subclasse para exibir o formato do arquivo
    def exibir_formato(self):
        print(f"Formato do Arquivo: {self.formato_arquivo}")
        print("-" * 30)

# Criando objetos da classe Livro e LivroDigital
livro1 = Livro("1984", "George Orwell", 1949)
livro2 = Livro("Dom Casmurro", "Machado de Assis", 1899)
livro_digital = LivroDigital("Python para Iniciantes", "Jo√£o Silva", 2021, "PDF")

# Exibindo dados dos livros
livro1.exibir_dados()
livro2.exibir_dados()

# Usando os m√©todos de a√ß√£o
livro1.emprestar()  # Emprestando livro1
livro2.devolver()   # Tentando devolver livro2 (n√£o emprestado)

# Exibindo dados do livro digital
livro_digital.exibir_dados()  # Exibindo dados do livro digital
livro_digital.exibir_formato()  # Exibindo o formato do arquivo

# Emprestando e devolvendo o livro digital
livro_digital.emprestar()  # Emprestando livro digital
livro_digital.devolver()   # Devolvendo livro digital
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).
- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python
# Classe Livro com encapsulamento e abstra√ß√£o
class Livro:
    def __init__(self, titulo, autor, ano_publicacao, preco):
        self.titulo = titulo
        self.autor = autor
        self.ano_publicacao = ano_publicacao
        self.__preco = preco  # Atributo privado

    # M√©todo para exibir informa√ß√µes sobre o livro
    def exibir_dados(self):
        print(f"T√≠tulo: {self.titulo}")
        print(f"Autor: {self.autor}")
        print(f"Ano de Publica√ß√£o: {self.ano_publicacao}")
        print(f"Pre√ßo: R${self.get_preco():.2f}")
        print("-" * 30)
    
    # M√©todo de a√ß√£o para emprestar o livro
    def emprestar(self):
        print(f"O livro '{self.titulo}' foi emprestado.")

    # M√©todo para calcular o pre√ßo com um desconto (l√≥gica interna)
    def aplicar_desconto(self, percentual_desconto):
        if percentual_desconto < 0 or percentual_desconto > 100:
            print("Percentual de desconto inv√°lido!")
        else:
            desconto = (self.__preco * percentual_desconto) / 100
            novo_preco = self.__preco - desconto
            print(f"Novo pre√ßo do livro '{self.titulo}' ap√≥s {percentual_desconto}% de desconto: R${novo_preco:.2f}")
            self.__preco = novo_preco

    # M√©todo get para acessar o atributo privado __preco
    def get_preco(self):
        return self.__preco

    # M√©todo set para alterar o valor do atributo privado __preco
    def set_preco(self, novo_preco):
        if novo_preco < 0:
            print("Pre√ßo n√£o pode ser negativo!")
        else:
            self.__preco = novo_preco
            print(f"Pre√ßo do livro '{self.titulo}' alterado para R${novo_preco:.2f}")

# Subclasse LivroDigital
class LivroDigital(Livro):
    def __init__(self, titulo, autor, ano_publicacao, preco, formato_arquivo):
        super().__init__(titulo, autor, ano_publicacao, preco)
        self.formato_arquivo = formato_arquivo

    def exibir_formato(self):
        print(f"Formato do Arquivo: {self.formato_arquivo}")
        print("-" * 30)

# Criando objetos
livro1 = Livro("1984", "George Orwell", 1949, 39.90)
livro2 = Livro("Dom Casmurro", "Machado de Assis", 1899, 29.90)
livro_digital = LivroDigital("Python para Iniciantes", "Jo√£o Silva", 2021, 49.90, "PDF")

# Exibindo dados dos livros
livro1.exibir_dados()
livro2.exibir_dados()

# Aplicando desconto no livro
livro1.aplicar_desconto(10)  # Aplicando 10% de desconto no pre√ßo do livro

# Alterando o pre√ßo do livro diretamente com o m√©todo set
livro2.set_preco(25.00)  # Alterando o pre√ßo para R$25.00

# Exibindo novamente os dados ap√≥s altera√ß√µes
livro1.exibir_dados()
livro2.exibir_dados()

# Exibindo dados do livro digital
livro_digital.exibir_dados()  # Exibindo os dados do livro digital
livro_digital.exibir_formato()  # Exibindo o formato do arquivo
```
